```python
%pylab inline
import pandas
import seaborn
```

# Load CSV file into memory

```python
data = pandas.read_csv('Desktop/uber-raw-data-apr14.csv')
```

```python
data.tail()
```

# Convert datetime and add some useful columns

```python
data['Date/Time'] = data['Date/Time'].map(pandas.to_datetime)
```

```python
data.head()
```

```python
#day of month
def get_dom(dt):
    return dt.day

data['dom'] = data['Date/Time'].map(get_dom)
```

```python
data.tail()
```

```python
def get_weekday(dt):
    return dt.weekday()

data['weekday'] = data['Date/Time'].map(get_weekday)

def get_hour(dt):
    return dt.hour

data['hour'] = data['Date/Time'].map(get_hour)

data.tail()
```

# Analysis

## Analyzing the Day of Month

```python
hist(data.dom, bins = 30, rwidth = 0.8, range = (0.5, 30.5))
xlabel('date of the month')
ylabel('frequency')
title('Fequecny by DoM - uber - Apr 2014')
```

```python
def count_rows(rows):
    return len(rows)

by_date = data.groupby('dom').apply(count_rows)
by_date
```

```python
plot(by_date)
```

```python
by_date_sorted = by_date.sort_values()
by_date_sorted
```

```python
figure(figsize = (10,10))
bar(range(1,31), by_date_sorted)
xticks(range(1,31), by_date_sorted.index)
xlabel('date of the month')
ylabel('frequency')
title('Fequecny by DoM - uber - Apr 2014')
;
```

## Analyzing the Hour

```python
hist(data.hour, bins = 24, rwidth = 0.8, range=(.5,24))
xlabel('hour of the day')
ylabel('frequency')
title('Fequecny by hour - uber - Apr 2014')
```

## Analyzing the Weekday

```python
hist(data.weekday, bins = 7, range = (-.5, 6.5), rwidth = 0.8, color = '#AA6666', alpha=.8)
xticks(range(7), 'Mon Tue Wed Thu Fri Sat Sun'.split())
xlabel('day of the week')
ylabel('frequency')
title('Fequecny by Weekday - uber - Apr 2014')
```

## Cross analysis (hour, weekday)

```python
by_cross = data.groupby('weekday hour'.split()).apply(count_rows).unstack()
by_cross
```

```python
seaborn.heatmap(by_cross)
```

## Analyzing by Lat and Lon

```python
hist(data['Lat'], bins = 100, range = (40.5, 41))
;
```

```python
figure(figsize = (8,8))
hist(data['Lon'], bins = 100, range = (-74.1, -73.9))
;
```

```python
figure(figsize = (10,10))
hist(data['Lon'], bins = 100, range = (-74.1, -73.9), color = 'g', alpha = .5, label = 'longitude')
legend(loc = 'upper left')
twiny()
hist(data['Lat'], bins = 100, range = (40.5, 41), color = 'r', alpha = .5, label = 'latitude')
legend(loc = 'best')
grid()
;
```

```python
figure(figsize = (10,10))
plot(data['Lon'], data['Lat'], '.',ms = 1, alpha = .5)
xlim(-74.2,-73.7)
ylim(40.7,41)
```
